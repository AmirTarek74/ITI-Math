```python
import numpy as np
import scipy as sp
from scipy.stats import randint, norm
import matplotlib.pyplot as plt
import pandas as pd
```

### Uniform discrete random variable.

p(x) = `1/high-low` where `high` is largest possible outcome and `low` is smallest possible outcome.
and x must belongs to {low,..., high-1}.

```python
#assume dice has been thrown 
low = 1   # smallest possible outcome
high = 7   # largest possible outcome has been incremented by 1 to include 6 in the output
tested_number_probablity = 4
print(f"Probabilty of {tested_number_probablity} when I throw a fair dice = {randint.pmf(tested_number_probablity, low, high):.4f}")
```

### Normal Distrbuation 

here we are looking for `PDF` or `CDF` for a normal distrbuation with specific $\mu$ and $\sigma$.

$f(x) =\frac {exp(\frac {-x{2}} {2})} {\sigma \sqrt{2\pi}} $ where $ x = \frac {z-\mu} {\sigma} $ and `z` is the value to calculate `PDF` or `CDF` at.

```python
sigma = 1
mu = 0
x = np.linspace(-10,10)
pdf = norm.pdf(x,mu,sigma)

plt.plot(x, pdf)
plt.xlabel('Value')
plt.ylabel('Distrubation')
plt.show()
```

```python
cdf = norm.cdf(x,mu,sigma)

plt.plot(x, cdf)
plt.xlabel('Value')
plt.ylabel('Probabilty')
plt.show()
```
